```python
def nullAnalisys(x):

    null = base[x].isna().sum()
    percent = null*100/len(base)

    return print('Valores nulos em', x, null, '---', f'{percent:.2f}%')
```

```python
nullAnalisys('Military interference in rule of law and politics')
nullAnalisys('Reliability of police')
nullAnalisys('Ownership of banks')
```
