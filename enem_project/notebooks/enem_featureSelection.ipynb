```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
```

```python
base = pd.read_csv('../data/enem_classificado.csv')
base.head()
```

```python
X = base.iloc[:,0:10].values
y = base.iloc[:,10].values
```

```python
modelo = LogisticRegression(max_iter = 2000, n_jobs = -1)
```

```python
seletor = RFE(modelo)
seletor = seletor.fit(X, y)
```

```python
features = seletor.get_support(indices = True)
features
```

```python
base.iloc[:,features]
```

```python
seletor.ranking_
```
