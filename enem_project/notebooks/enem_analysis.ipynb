```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
base = pd.read_csv('../data/microdados_enem_2021.csv', sep = ';', encoding = 'ISO-8859-1')
pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', 100)
base.head()
```

```python
cols = ['TP_FAIXA_ETARIA', 'TP_ESTADO_CIVIL', 
        'TP_ESCOLA', 'Q001', 'Q002', 'Q003', 
        'Q004', 'Q006', 'Q024', 'Q025', 
        'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 
        'NU_NOTA_MT', 'NU_NOTA_REDACAO']
```

```python
base = base[cols]
base.head()
```

```python
base.shape
```

```python
base.isna().sum()
```

```python
base.dropna(inplace = True)
base.isna().sum()
```

```python
base.shape
```

```python
base['MEDIA_GERAL'] = (base['NU_NOTA_CN'] + base['NU_NOTA_CH'] +
                       base['NU_NOTA_LC'] + base['NU_NOTA_MT'] + 
                       base['NU_NOTA_REDACAO']) / 5
base.head()
```

```python
base['MEDIA_GERAL'].describe()
```

```python
#Verificando se existe m√©dia geralcom nota 0, que possivelmente s√£o de pessoas que n√£o fizeram as provas
base.loc[base['MEDIA_GERAL'] == 0]
```

```python
#Modificando a base de dados excluindo as m√©dias 0
base = base.loc[base['MEDIA_GERAL'] != 0]
base.shape
```

```python
base = base.drop(['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
                'NU_NOTA_MT', 'NU_NOTA_REDACAO'], axis = 1)
base.shape
```

```python
base['MEDIA_GERAL'].describe()
```

```python
#Verificando o tipo de escola exclu√≠ndo as pessoas que n√£o responderam
base.loc[base['TP_ESCOLA'] != 1]
```

```python
'''
Aqui iremos modificar o dataframe e excluindo o tipo
de escola 1, cujo valor significa que as pessoas que
realizaram o exame n√£o responderam se conclu√≠ram
o ensino m√©dio em escola p√∫blica ou privada.
'''
base = base.loc[base['TP_ESCOLA'] != 1]
base.shape
```

```python
label_encoder = LabelEncoder()
columns = ['Q001', 'Q002',	'Q003',	'Q004',	'Q006',	'Q024',	'Q025']
for i in range(len(columns)):
    base[columns[i]] = label_encoder.fit_transform(base[columns[i]])
base.head()
```

```python
#Salvando a base de dados modificada com as m√©dias pontuadas
#base.to_csv('enem_pontuado.csv', index = False)
```

```python
#Criando classes e agrupando de forma igualmente distribu√≠da com o m√©todo cut
pd.cut(base.MEDIA_GERAL, bins = [0,500,700,1000], labels = ['baixa', 'm√©dia', 'alta']).value_counts()
'''
Al√©m do m√©todo cut, temos tamb√©m o qcut, que j√° faz a divis√£o autom√°tica dos dados,
balanceando as classes deixando-as igualmente distribu√≠das, por√©m esse m√©todo causou
orverlap nos dados, onde as classes ficaram com as fronteiras mal definididas causando
assim um ru√≠do e baixo desempenho do modelo.
'''
```

```python
#Aplicando a divis√£o realizada na nova coluna NOTA
base['NOTA'] = pd.cut(base.MEDIA_GERAL, bins = [0,500,700,1000], labels = ['baixa', 'm√©dia', 'alta'])
base.head()
```

```python
base.drop(['MEDIA_GERAL'], axis = 1, inplace = True)
base
```
