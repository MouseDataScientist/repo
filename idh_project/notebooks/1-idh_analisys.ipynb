```python
def nullAnalisys(x):
    num = base[x].isna().sum()
    percent = num*100/len(base)
    return print('Nulos em', x.casefold(), '=', num, '\n', f'{percent:.2f}%', '\n')
```

```python
indexes = list(base.columns)
count = int(0)
for i in range(len(base.columns)):
    x = indexes[i]
    num = base[x].isna().sum()
    if num > 0:
        nullAnalisys(indexes[i])
    else: 
        count = count+1
        if count == len(base.columns):
            print('Não há valores nulos na base de dados!')
```
