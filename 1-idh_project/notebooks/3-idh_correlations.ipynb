```python
import pandas as pd
import numpy as np
```

```python
base = pd.read_csv('../data/economicdata2020v2.csv', sep=',')
base.head()
```

```python
c_pearson = base.corr(method='pearson').reset_index()
c_pearson
```

```python
c_pearson = c_pearson[['index', 'human development index']]
c_pearson
```

```python
c_pearson = c_pearson.loc[(c_pearson['human development index'] > 0.5) & (c_pearson['human development index'] < 1)]
c_pearson
```

```python
c_pearson.shape
```

```python
countries = np.array(['countries'])
indexies = list(c_pearson['index'])
countries = np.append(countries, [indexies])
indexies = np.append(countries, ['human development index','categories'])
```

```python
indexies
```

```python
base = base[indexies]
```
