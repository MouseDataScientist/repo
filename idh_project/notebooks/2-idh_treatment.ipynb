```python
import pandas as pd
import numpy as np
```

```python
base = pd.read_csv('../data/economicdata2020.csv', sep=',')
#pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
base.head()
```

```python
base.shape
```

```python
base = base.drop(['data', 'data.1', 'data.2'], axis=1)
```

```python
base.shape
```

```python
# Criação de classe categória para posteriormente nos auxiliar na análise gráfica
base['Categories'] = pd.cut(base['Human Development Index'], bins=[0.350,0.554,0.699,0.799,1.000], labels=['low','medium','high','very_high'])
```

```python
base['Categories'].value_counts()
```

```python
base.describe()
```

```python
# Código para analisar e criar uma lista dos atributos que possuem valores nulos 
baseIndexes = list(base.columns)
nullIndexes = []
count = int(0)
for i in range(len(base.columns)):
    x = baseIndexes[i]
    num = base[x].isna().sum()
    if num > 0:
        nullIndexes = np.append(nullIndexes, base.columns[i])
        percent = num*100/len(base[x])
        print('Nulos em', x.casefold(), '=', num, '\n', f'{percent:.2f}%', '\n')
    else: 
        count = count+1
        if count == len(base.columns):
            print('Não há valores nulos na base de dados!')
```

```python
# Com a lista de atributos que possuem valores nulos criada, podemos aplicar a média para esses valores
for i in range(len(nullIndexes)):
    x = nullIndexes[i]
    mean = base[x].mean()
    base[x].fillna(mean, inplace=True)
```

```python
base.isna().sum()
```

```python
# Retirando espaços e padronizando a nomenclatura dos atributos em minúsculas
base.rename(columns={' Judicial independence':'Judicial independence',
            ' Regulatory trade barriers':'Regulatory trade barriers',
            }, inplace=True)

for i in range(len(baseIndexes)):
    if baseIndexes[i].casefold() == base.columns[i].casefold():
        base.rename(columns={base.columns[i]:base.columns[i].casefold()}, inplace=True)
```
