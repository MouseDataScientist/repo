```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
base = pd.read_csv('../data/microdados_enem2021.csv', sep=';', encoding='ISO-8859-1')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 76)
base.head()
```

```python
base.shape
```

```python
"""
Criando uma lista de √≠dices para selecionar quais os √≠ndices desejados 
para o projeto atrav√©s do n√∫mero do √≠ndice.
"""
indexes = pd.DataFrame(list(base.columns))
indexes
```

```python
indexes = list(base.columns[[2,4,9,51,52,53,54,56,74,75,31,32,33,34,50]])
indexes
```

```python
base = base[indexes]
base.head()
```

```python
base.shape
```

```python
base.isna().sum()
```

```python
base.dropna(inplace = True)
base.isna().sum()
```

```python
base.shape
```

```python
base['MEDIA_GERAL'] = (base['NU_NOTA_CN'] + base['NU_NOTA_CH'] +
                       base['NU_NOTA_LC'] + base['NU_NOTA_MT'] + 
                       base['NU_NOTA_REDACAO']) / 5
base.head()
```

```python
base.drop(['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO'], axis = 1, inplace = True)
```

```python
base['MEDIA_GERAL'].describe()
```

```python
#Verificando se existe m√©dia geral com nota 0
base.loc[base['MEDIA_GERAL'] == 0]
```

```python
base = base.loc[base['MEDIA_GERAL'] != 0]
base.shape
```

```python
base['MEDIA_GERAL'].describe()
```

```python
#Verificando o tipo de escola exclu√≠ndo as pessoas que n√£o responderam
base.loc[base['TP_ESCOLA'] != 1]
```

```python
base = base.loc[base['TP_ESCOLA'] != 1]
base.shape
```

```python
label_encoder = LabelEncoder()
columns = ['Q001', 'Q002',	'Q003',	'Q004',	'Q006',	'Q024',	'Q025']
for i in range(len(columns)):
    base[columns[i]] = label_encoder.fit_transform(base[columns[i]])
base.head()
```

```python
#Criando classes e agrupando de forma igualmente distribu√≠da com o m√©todo cut
pd.cut(base.MEDIA_GERAL, bins = [0,500,700,1000], labels = ['baixa', 'm√©dia', 'alta']).value_counts()
"""
Al√©m do m√©todo cut, temos tamb√©m o qcut, que j√° faz a divis√£o autom√°tica dos dados,
balanceando as classes deixando-as igualmente distribu√≠das, por√©m esse m√©todo causou
orverlap nos dados, onde as classes ficaram com as fronteiras mal definididas causando
assim um ru√≠do e baixo desempenho do modelo.
"""
```

```python
#Aplicando a divis√£o realizada na nova coluna NOTA
base['NOTA'] = pd.cut(base.MEDIA_GERAL, bins = [0,500,700,1000], labels = ['baixa', 'm√©dia', 'alta'])
base.head()
```

```python
base.drop(['MEDIA_GERAL'], axis=1, inplace = True)
base.head()
```
