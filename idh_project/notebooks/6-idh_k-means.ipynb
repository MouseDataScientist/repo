```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
base = pd.read_csv('../data/economicdata2020v3.csv', sep=',')
base.head()
```

```python
base = base.drop(['human development index'], axis=1)
```

```python
class_orderly = pd.CategoricalDtype(categories = ['very_high', 'high', 'medium', 'low'], ordered=True)
base['categories'] = base['categories'].astype(class_orderly)
```

```python
base.sort_values('categories', ascending=True, inplace=True)
base
```

```python
base.groupby(['categories']).size()
```

```python
x = base.iloc[:, 1:19].values
y = base.iloc[:, 19]
```

```python
print(x)
print(y)
```

```python
y.replace({'very_high':3, 'high':2, 'medium':1, 'low':0}, inplace=True)
```

```python
groups, quantities = np.unique(y, return_counts=True)
```

```python
groups
```

```python
quantities
```

```python
#Definindo o nÃºmero de clusters(grupos)
cluster = KMeans(n_clusters=4)
cluster.fit(x)
```

```python
#Visualizando os centros utilizados pelo agrupador
centroids = cluster.cluster_centers_ 
centroids
```

```python
#Visualizando o agrupamento realizado
grouping = cluster.labels_
grouping
```

```python
#Contagem de grupos e quantidades de cada grupo
groups2, quantities2 = np.unique(grouping, return_counts=True)
```

```python
#Quantidade de grupos
groups2
```

```python
#Quantidades dentro de cada grupo
print(quantities)
print(quantities2)
```

```python
#Analisando a matriz de confusÃ£o 
results = confusion_matrix(y, grouping)
results
```

```python
#Verificando a performance de acerto do agrupamento K-means
rate = accuracy_score(y, grouping)
rate
```
