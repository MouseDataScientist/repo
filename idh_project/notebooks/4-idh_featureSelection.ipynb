```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
```

```python
base = pd.read_csv('../data/economicdata2020v2.csv')
pd.set_option('display.max_columns', None)
base
```

```python
df = base.drop(['countries', 'human development index'], axis=1)
df.head()
```

```python
X = df.iloc[:,0:43].values
y = df.iloc[:,43].values
print(X,'\n')
print(y)
```

```python
model = LogisticRegression(max_iter=2000, n_jobs=-1)
```

```python
selector = RFE(model)
selector = selector.fit(X, y)
```

```python
selector.ranking_
```

```python
features = selector.get_support(indices=True)
features
```

```python
features = np.append(features, 44)
features = np.append(features, 45)
```

```python
features
```

```python
base = base.iloc[:,features]
```
