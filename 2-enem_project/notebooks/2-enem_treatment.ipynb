```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
base = pd.read_csv('../data/microdados_enem2021.csv', sep=';', encoding='ISO-8859-1')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 76)
base.head()
```

```python
base.shape
```

```python
'''
Criando uma lista de ídices para selecionar quais os índices desejados 
para o projeto através do número do índice.
'''
indexes = pd.DataFrame(list(base.columns))
indexes
```

```python
indexes = list(base.columns[[2,9,10,56,72,74,75,31,32,33,34,50]])
indexes
```

```python
base = base[indexes]
base
```

```python
base.isna().sum()
```

```python
# Em TP_ENSINO temos mais de 1 milhão de registtros nulos, sendo assim, ao invés
# de simplesmente nos desfazermos de todos esses registros só por causa de 1 atributito
# nulo, vamos criar uma variável "unknown" no qual receberá o valor zero.
base['TP_ENSINO'].fillna(float(0), inplace=True)
```

```python
base
```

```python
# Drop nas colunas "Q001".
base.dropna(subset=['Q006'], inplace=True)
```

```python
# Observando a nossa base de dados percebemos que nem sempre os valores NaN são de participantes 
# que não realizaram o exame, na verdade alguns optaram por não realizar todas as matérias, sendo 
# assim, para os valores NaN das matérias vamos atribuir a nota 0.
materias = ['NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT','NU_NOTA_REDACAO']
for i in range(len(materias)):
    base[materias[i]].fillna(float(0), inplace=True)
```

```python
base
```

```python
# Criando um atributo que receberá a média das nota das matérias. 
base['MEDIA_GERAL'] = (base['NU_NOTA_CN'] + base['NU_NOTA_CH'] +
                       base['NU_NOTA_LC'] + base['NU_NOTA_MT'] + 
                       base['NU_NOTA_REDACAO']) / 5
base.head()
```

```python
# Eliminando as notas indiciduais.
base.drop(['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO'], axis = 1, inplace = True)
```

```python
base['MEDIA_GERAL'].describe()
```

```python
# Verificando qauntas notas 0 existem em nossa base de dados, pois se forem poucas podemos 
# substituir com a mediana sem distorcer muito nossos dados.
base.loc[base['MEDIA_GERAL'] == 0]
```

```python
# Podemos verificar que muitos participantes responderam questionários socioeconômicos porém não
# realizaram a avaliação de NENHUMA matéria, sendo assim vamos retirar esses dados porque não
# podemos verificar a relação deles com a média geral. 
base = base.loc[base['MEDIA_GERAL']!=0]
base
```

```python
base.isna().sum()
```

```python
# Criando classes e agrupando essas classes com o método cut, para que posteriormente
# possamos destacar essas classes nos gráficos seaborn.
pd.cut(base.MEDIA_GERAL, bins = [0,500,700,1000], labels = ['baixa', 'media', 'alta']).value_counts()
```

```python
base = base.dropna()
```

```python
# Aplicando a divisão realizada na nova coluna NOTA.
base['NOTA'] = pd.cut(base.MEDIA_GERAL, bins = [0,500,700,1000], labels = ['baixa', 'media', 'alta'])
base.head()
```

```python
base.drop(['MEDIA_GERAL'], axis=1, inplace=True)
```

```python
label_encoder = LabelEncoder()
lbl_columns = ['Q006','Q022','Q024','Q025']
```

```python
for i in range(len(lbl_columns)):
    base[lbl_columns[i]] = label_encoder.fit_transform(base[lbl_columns[i]])
```

```python
base = base.astype({'TP_ENSINO':int})
base
```

```python
base.rename(columns = {'Q006':'RENDA_FAM', 'Q022':'CELULAR', 'Q024':'COMPUTADOR', 'Q025':'INTERNET'}, inplace = True)
base
```
